# Secant method

```python
def Hord (A, B, f):
  i = 0
  C=(A*f(B)-B*f(A))/(f(B)-f(A))
  while abs(f(C))>=epsilon:
    i=i+1
    print (str(i) + ") [A, B] = [" + str(A) + ", " + str(B) + "]")
    C=(A*f(B)-B*f(A))/(f(B)-f(A))
    if f(A)*f(C)<=0:
      B=C
      root = (A+B)/2
    elif f(C)*f(B)<=0:
      A=C
      root = (A+B)/2
  return root
```
