## 3. Лістинг програми уточнення коренів за методами бісекції, хорд, дотичних (вхідними даними для цієї програми є координати проміжків $[ a_i, b_i ]$ та коефіцієнти поліному) та результати дії програми. На кожній ітерації методу слід виводити такі дані: номер ітерації, наближене значення кореня, критерій завершення ітерацій.

### Метод бісекцій (поділу навпіл)


```
def biSect (A, B, f):
  i = 0
  while abs((A-B)/2)>=0.00001:
    i=i+1
    print (str(i) + ") [A, B] = [" + str(A) + ", " + str(B) + "]")
    C=(A+B)/2
    if f(A)*f(C)<=0:
      B=C
      root = (A+B)/2
    elif f(C)*f(B)<=0:
      A=C
      root = (A+B)/2
  return root
```

результат:

```
print("the root1 is ", biSect (arr1[0], arr1[1], f))
print("the root2 is ", biSect (arr2[0], arr2[1], f))
print("the root3 is ", biSect (arr3[0], arr3[1], f))
```

### Метод Ньютона

```
def Newton(x_k0):
  def f(x):
    return x**5-3*x**4+7*x**2-3
  def df(x):
    return 5*x**4-12*x**3+14*x

  i=0
  x_k=x_k0
  while abs(f(x_k))>0.00001:
    x_k=x_k-f(x_k)/df(x_k)
    i=i+1
    print(str(i)+")	",x_k)
  
  print("root is ", x_k)
```

результат:

```
Newton(arr1[0])
Newton(arr2[0])
Newton(arr3[0])
```

### Метод хорд

```
def Hord (A, B, f):
  i = 0
  C=(A*f(B)-B*f(A))/(f(B)-f(A))
  while abs(f(C))>=0.00001:
    i=i+1
    print (str(i) + ") [A, B] = [" + str(A) + ", " + str(B) + "]")
    C=(A*f(B)-B*f(A))/(f(B)-f(A))
    if f(A)*f(C)<=0:
      B=C
      root = (A+B)/2
    elif f(C)*f(B)<=0:
      A=C
      root = (A+B)/2
  return root
```

результат:


```
print("the root1 is ", Hord (arr1[0], arr1[1], f))
print("the root2 is ", Hord (arr2[0], arr2[1], f))
print("the root3 is ", Hord (arr3[0], arr3[1], f))
```
